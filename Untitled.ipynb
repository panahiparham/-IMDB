```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
```

### impoting data

```python
_title_ratings = pd.read_csv(".\\IMDB_files\\title.ratings.tsv", sep='\t')
```

```python
_title_basics = pd.read_csv(".\\IMDB_files\\title.basics.tsv", sep='\t')
```

```python
_title_ratings.head()
```

```python
_title_basics.head()
```

### grab titles with more than 1000 votes

```python
title_ratings = _title_ratings[_title_ratings['numVotes'] > 1000]
```

```python
title_ratings.describe()
```

### grab movie data and drop Nan values

```python
title_basics = _title_basics[_title_basics['titleType'] == 'movie'][['tconst', 'primaryTitle', 'startYear', 'runtimeMinutes', 'genres']]
```

```python
title_basics = title_basics.replace('\\N', np.nan).dropna()
```

```python
title_basics
```

### merge two dataframes

```python
movie_data = pd.merge(title_basics, title_ratings, how='inner', on=['tconst'])
movie_data.reset_index()
```

```python
# fixing start year
movie_data['startYear'] = movie_data['startYear'].astype('int')
```

```python
# filter year : after 1990
movie_data = movie_data[movie_data['startYear'] >  1990]
```

```python
# fixing runtime
movie_data['runtimeMinutes'] = movie_data['runtimeMinutes'].astype('int')
```

```python
# rating per vote feature
movie_data['rating/vote'] = movie_data['averageRating'] / movie_data['numVotes']
```
